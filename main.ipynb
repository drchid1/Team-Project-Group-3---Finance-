```python
# Importing of Python Pandas Module

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Creating Data frame for each of the 4 selcted stocks from the CSV files
# Note: the stock ticker symbol preceeds the df notation eg: wmf_df is the Wallmart Data Frame
aci_df = pd.read_csv("Data/ACI.csv")
cost_df = pd.read_csv("Data/COST (2).csv")
tgt_df = pd.read_csv("Data/TGT.csv")
wmt_df = pd.read_csv("Data/WMT.csv")
```

```python
cost_df.head()
```

```python
plt.bar(cost_df['Date'], cost_df['Volume'])
plt.show()
```

```python
def plot_stock_vol(ticker, ticker_df):
    plt.bar(ticker_df['Date'], (ticker_df['Volume']), width=0.7)
    plt.title(f"{ticker} Stock Volume Change over 5 year")
    plt.xlabel("Time (Months)")
    plt.ylabel("Stock Volume Change")
    plt.xticks(rotation=90)
    plt.xticks(fontsize=8)
    plt.tight_layout()
    plt.savefig(f"output/{ticker} Stock Vol Plot.png")
```

```python
plot_stock_vol(ticker = "COST",
               ticker_df = cost_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read dividend data from CSV files
dividend_data_cost = pd.read_csv('COST Dividend Data.csv')
dividend_data_wmt = pd.read_csv('WMT Dividend Data.csv')
dividend_data_tgt = pd.read_csv('TGT Dividend Data.csv')
dividend_data_aci = pd.read_csv('ACI Dividend Data.csv')

# Add a 'Symbol' column to each DataFrame
dividend_data_cost['Symbol'] = 'COST'
dividend_data_wmt['Symbol'] = 'WMT'
dividend_data_tgt['Symbol'] = 'TGT'
dividend_data_aci['Symbol'] = 'ACI'

# Combine the DataFrames into a single DataFrame
combined_dividend_data = pd.concat([dividend_data_cost, dividend_data_wmt, dividend_data_tgt, dividend_data_aci])

# Group the combined DataFrame by symbol and count the number of dividend payouts for each company
dividend_counts = combined_dividend_data.groupby('Symbol').size().reset_index(name='Dividend_Count')

# Define colors for bars
colors = ['red', 'skyblue', 'green', 'orange']

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(dividend_counts['Symbol'], dividend_counts['Dividend_Count'], color=colors)

# Add labels and title
plt.ylabel('Number of Dividend Payments')
plt.title('Number of Dividend Payments for Each Company')

# Show plot
plt.show()


```

```python
#csv to df
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#stockmarket average price
stockmonth_df = pd.read_csv("data/Stock_monthly.csv")
```

```python
stockmonth_df["Average"] = (stockmonth_df["High"] + stockmonth_df["Low"])/2
```

```python
dates = stockmonth_df["Date"]
stockhigh = stockmonth_df["High"]
stocklow = stockmonth_df["Low"]
stockaverage = stockmonth_df["Average"]
```

```python
sm = plt.plot(dates, stockaverage)
sm
```

```python
#stock market average percent change
stockmonth_df["Difference"] = stockmonth_df["Open"] - stockmonth_df["Close"]
stockmonth_df["Percent"] = (stockmonth_df["Difference"]/stockmonth_df["Open"])*100
stockpercent = stockmonth_df["Percent"]

smp = plt.plot(dates, stockpercent)
smp
```

```python
#all average price compared to stockmarket

fig, ax1 = plt.subplots(figsize=(12, 12))
ax2 = ax1.twinx()
ax3 = ax1.twinx()

ax1.bar(dates, stockaverage, color="limegreen", alpha = 0.2)

ax2.plot(costdate, costave, color="red", label="Costco")
ax2.plot(tgtdate, tgtave, color="orangered", label="Target")

ax3.plot(acidate, aciave, color="dodgerblue", label="Albertsons")
ax3.plot(wmtdate, wmtave, color="navy", label="Walmart")

plt.title("Title")
plt.legend(loc="upper left")

ax1.set_xlabel("Time")
ax1.set_ylabel("$ Stockmarket Value", color="limegreen")
ax1.set_ylim(0, 16000)

ax2.set_ylabel("$larger", color="red")
ax2.tick_params(axis="y", labelcolor="red")
ax2.set_ylim(0, 800)

ax3.set_ylabel("$smaller", color="blue")
ax3.tick_params(axis="y", labelcolor="blue")
ax3.set_ylim(0, 80)


```

```python
# higher value supermarket stocks compared to stock market

fig, ax1 = plt.subplots(figsize=(12, 8))
ax2 = ax1.twinx()

ax1.bar(dates, stockaverage, color="limegreen", alpha = 0.2)

ax2.plot(costdate, costave, color="red", label="Costco")
ax2.plot(tgtdate, tgtave, color="orange", label="Target")

plt.title("Higher priced supermarket stock value compared to stock market")
plt.legend(loc="upper left")

ax1.set_xlabel("Time")
ax1.set_ylabel("$ Stockmarket Value", color="limegreen")
ax1.set_ylim(0, 16000)

ax2.set_ylabel("$ Stock Value", color="red")
ax2.tick_params(axis="y", labelcolor="red")
ax2.set_ylim(0, 800)

ax1.tick_params(rotation = 90)

plt.savefig("../output/highervalue.png")
```
